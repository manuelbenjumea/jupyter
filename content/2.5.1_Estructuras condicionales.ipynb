El  flujo de un programa  se refiere al orden en que se ejecutan las instrucciones. Hasta este momento, todas las instrucciones se ejecutan de manera secuencial. Es decir, una tras otra en el orden en que están escritas en el programa. Las distintas estructuras de programación (ej. Selección, repetición) permiten que el flujo sea modificado para responder a la interacción con el usuario o a las necesidades del problema que se espera resolver.


Considere un programa que le pregunte al usuario el tiempo en minutos que tarda haciendo una tarea y retorne valor que deberá pagarse asumiendo que se paga $\$5000$ por hora completa de labor, y que la fracción de la hora se paga a $\$80$ por minuto. Note la manera como las instruccciones se ejecutan de manera secuencial. 


````python
entrada  = input("Ingrese el número de minutos requeridos para la tarea: ")
minutos = int(entrada) # Convertir a entero
horas = minutos // 60 # División entera, obtine el número de horas
restante = minutos - horas * 60 # Obtiene el número restante de minutos. 
# Tambien puede obtenerse con el operador modulo
restante = minutos % 60 
pago = horas*5000 + restante*80 # Calcula el pago
print("La tarea toma ", horas, " y ", restante, " minutos. con un valor de: ", pago)
````

Considere ahora el caso en el cual se paga $\$5000$ por hora completa de labor. Si se trabaja una fracción de una hora que sea superior a la mitad de dicha hora deberá pagarse el valor completo ($\$5000$), de lo contrario deberá pagarse a $\$80$ por minuto. **¿Cómo debería modificarse el flujo?**

En estos casos se hace uso de las estructuras condicionales o selectivas en las cuales se evalúa una condición y en función del resultado de la misma se realiza una de varias opciones. Estas estructuras pueden ser:

* simples,
* dobles,
* múltiples.

## <font color='157699'> **Estructura condicional simple** </font>

En esta estructura se ejecuta una determinada instrucción cuando se cumple una condición. Es decir, esta estructura evalúa la condición y:
* si la condición es verdadera, entonces ejecuta la instrucción o conjunto de instrucciones,
* si la condición es falsa, no ejecuta la instrucción o conjunto de instrucciones.




![title](https://docs.google.com/uc?export=download&id=1HN_RZYxdlyZ1UAJ2P7eklplsax3AnoSC)

La sintaxis para escribir este tipo de estructuras cambia dependiendo del lenguaje pero en general podrá verse de la siguiente forma:
```Python
If (Condición) Then
    Instrucciones a realizar si la condición es verdadera
End If
```

<font color='157699'> **Note que** hemos usado dos formas diferentes de describir un algoritmo. La primera de ellas a través de un *diagrama de flujo* y la segunda a través de un *pseudocódigo*. Más adelante, ahondaremos en la importancia de estas dos representaciones </font>

## <font color='157699'> **Estructura condicional doble** </font>

Extiende la estructura simple descrita anteriormente. La estructura condicional doble permite elegir entre dos opciones
o alternativas posibles, en función del cumplimiento o no de una determinada condición. Si la condición es verdadera, se ejecuta una acción y, si es falsa, se ejecuta otra acción diferente

![title](https://docs.google.com/uc?export=download&id=1uyTRi4MfzOV-VGVsqGvYtJt1F1M86CKd)

La sintaxis para escribir este tipo de estructuras cambia dependiendo del lenguaje pero en general podrá verse de la siguiente forma:
```Python
If (Condición) Then
    Instrucciones a realizar si la condición es verdadera
Else
    Instrucciones a realizar si la condición es falsa
End If
```



Considere el siguiente programa de ejemplo:
Usted debe ingresar el valor de la compra de un cliente. Si dicha compra es superior a $\$20.000$ el cliente tendrá derecho a un descuento del 15 $\%$, en caso contrario el descuento será únicamente del 5 $\%$. El siguiente diagrama describe el programa que se desea implementar.



![title](https://docs.google.com/uc?export=download&id=1V-EC2j7c77y1N5M1Sja52FFMoT4Pw6VP)



Observe el código del programa que implementa el algoritmo



````python
compra = input("introduzca el valor total de la compra:") # pregunta el valor de la compra
valor = int(compra) # obtiene el valor como entero
if valor >= 20000: # Verifica si se cumple la condición
    print ("Usted tiene derecho a un descuento del 15%")
    valor = 0.85*valor
else:
    print ("Usted tiene derecho a un descuento del 5%")
    valor = 0.95*valor
print("El valor de su compra es ", valor)
````

## <font color='157699'> **Estructura condicional anidada** </font>

Las estructuras de selección no necesariamente implican la selección de una de dos alternativas, es posible también utilizarlas  para diseñar estructuras de selección que contengan más alternativas. Por ejemplo, una estructura condicional puede contener otra estructura condicional, y esta estructura puede contener otra, y así sucesivamente cualquier número de veces

![title](https://docs.google.com/uc?export=download&id=1aBf--Q1P48_h8EA-BAypq1iiumn63AEw)

La sintaxis para escribir este tipo de estructuras cambia dependiendo del lenguaje pero en general podrá verse de la siguiente forma:

```Python
If (Condición) Then
    Instrucciones a realizar si la condición es verdadera
    If (Condición) Then
        Instrucciones a realizar si la condición es verdadera    
    Else
        Instrucciones a realizar si la condición es falsa
    End If
Else
    Instrucciones a realizar si la condición es falsa
End If
```

Considere el siguiente ejemplo:
Usted debe ingresar el valor de la compra de un cliente. Si dicha compra es inferior a $\$20.000$ el cliente tendra derecho a un descuento del 5$\%$, en caso contrario el descuento será del 15$\%$ si la compra esta entre $\$20.000$ y $\$50.000$ o del 35$\%$ si la compra es superior a $\$50.000$. El siguiente diagrama describe el programa que se desea implementar.

![title](https://docs.google.com/uc?export=download&id=1I3pI2D885685K4j55dG8Kory1Nria4wj)



````python
compra = input("introduzca el valor total de la compra:") # pregunta el valor de la compra
valor = int(compra) # obtiene el valor como entero
if valor < 20000: # Verifica si se cumple la condición
    print ("Usted tiene derecho a un descuento del 5%")
    valor = 0.95*valor
else:
    if valor <= 50000: # Verifica si se cumple la condición
        print ("Usted tiene derecho a un descuento del 15%")
        valor = 0.85*valor
    else:
        print ("Usted tiene derecho a un descuento del 35%")
        valor = 0.65*valor
print("El valor de su compra es ", valor)
````

Alternativamente, las condiciones pueden anidarse de modo que se verifique una de de las condiciones y no todas ellas usando la instrucción elseif. Al cumplirse una de dichas instrucciones las siguientes no se evaluarán. 
```
If (Condición) Then
    Instrucciones a realizar si la condición es verdadera
    Elseif (Condición) Then
        Instrucciones a realizar si la condición es verdadera    
    Elseif (Condición) Then
        Instrucciones a realizar si la condición es verdadera
Else
    Instrucciones a realizar si la condición es falsa
    Elseif (Condición) Then
        Instrucciones a realizar si la condición es verdadera
    Elseif (Condición) Then
        Instrucciones a realizar si la condición es verdadera
End If
```
